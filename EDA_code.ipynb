```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
#Since the ZomatoDataset also contain "Country Code" column, we'll combine 
final_df=pd.merge(df,df_country,on="Country Code", how='left')
```

```python
final_df.head(2)
```

```python
 #To check Data Types
final_df.dtypes
```

```python
#To check number of transactions made in each country
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#Pie Chart-Top three countries using zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: Maximum number of transactions are from India

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations: 
i. Ratings:equal to or between 4.5 and 4.9 ----> Excellent
ii. Ratings:equal to or between 4.0 and 4.4 ----> Very Good
iii. Ratings:equal to or between 3.5 and 3.9 ----> Good
iv. Ratings:equal to or between 2.5 and 3.4 ----> Average
v. Ratings:equal to or between 1.5 and 2.4 ----> Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

Observations:
i. Not Rated count is very high
ii. Maximum number of ratings are between 2.5 to 3.4

```python
#Count plot
sns.countplot(x="Rating color",hue='Rating color',data=ratings,palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
#Heatmap â€“ Correlation Between Numerical Features
plt.figure(figsize=(10, 6))
numeric_cols = df.select_dtypes(include=['int64', 'float64'])
correlation = numeric_cols.corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

Observation: The diagonal elements are 1.0, signifying that each variable is perfectly correlated with itself.

Observation: Most of the restaurants are average rated

```python
#Histogram-Distribution of Votes
plt.figure(figsize=(12, 6))
sns.histplot(df['Votes'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Votes')
plt.xlabel('Number of Votes')
plt.ylabel('Frequency')
plt.show()
```

```python
#Box Plot: Average Cost for Two by Price Range
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price range', y='Average Cost for two',hue='Price range',data=df, palette='Set2')
plt.title('Cost Distribution by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Cost for Two')
plt.show()
```

```python
#Box Plot: Votes by Rating Text
plt.figure(figsize=(10, 6))
sns.boxplot(x='Rating text', y='Votes',hue='Rating text',data=df, palette='muted')
plt.title('Votes by Rating Category')
plt.xticks(rotation=45)
plt.show()
```
